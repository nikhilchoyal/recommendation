```python
genres=[]
for i in temp.genres:
    for j in i:
        if j not in genres:
            genres.append(j)
df=pd.DataFrame(columns=genres,index=temp.title)
```

```python
d=[]
for i in temp.genres:
    df_n=[]
    for j in df:
        if j in i:
            df_n.append(1)
        else:
            df_n.append(0)
    d.append(df_n)
dft=pd.DataFrame(d,columns=genres,index=temp.title)
dft["studio"]=studio
dft["rating"]=df_mean
dft["vote_average"]=df_av_mean
dft["language"]=language
#dft["adult"]=adult
```

```python
dft=dft.drop_duplicates()
dft
```
